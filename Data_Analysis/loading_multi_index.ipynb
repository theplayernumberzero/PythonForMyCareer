```python
#pivot makes horizontally longer
data = pd.read_csv('happiness_report.csv')
data.pivot(index='Year', columns= 'Country Name') #set column and year (You can use multi indexing) (Columns boş kalabilir)

data.pivot(index=['Country Name'], columns= ['Year'], values=['Generosity', 'Social Support']) #choose value which you want to show
```

```python
#melt makes horizontally longer
data = pd.read_csv('happiness_report.csv')
data.melt(id_vars='Country Name', var_name='Metric', value_name= 'Value') #convert into variables and values
```

```python
#pivot table: summary
data = pd.read_csv('happiness_report.csv')
#values and indexs can be multi
data.pivot_table(values='Generosity', index='Country Name') #aggregated
data.pivot_table(values='Generosity', index='Country Name', aggfunc='sum') #aggregated
```
