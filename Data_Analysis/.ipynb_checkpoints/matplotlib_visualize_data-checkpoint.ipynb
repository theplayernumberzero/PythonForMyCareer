```python
# Plotting the first graph
months = ['Jan', 'Feb', 'March', 'April', 'May']
prices = [100, 29, 1000, 40, 11]
prices_2 = [1000, 209, 100, 400, 1100]

plt.plot(months, prices, label="Price 1")  # First plot
plt.plot(months, prices_2, label="Price 2")  # Second plot

# Labeling the axes and the plot
#plt.xlabel("Months")
#plt.ylabel("Price")
plt.legend(['2025','2026'])

plt.show()
```
